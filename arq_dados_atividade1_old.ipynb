```python
import re

def regra_1(frase):

  # 1. Verificar presença de verbos no presente do indicativo
  verbos_presente = re.findall(r"\w*\[ando|indo|iendo|amos|áis|ão|an|en|mos|em|ém|er|ar|ir]", frase)
  #print(verbos_presente)
  #print(frase.split())
  probabilidade_verbo = len(verbos_presente) / len(frase.split())  # Proporção de verbos no presente
  #print(probabilidade_verbo)

  # 2. Verificar presença de letras ou símbolos
  presenca_es = "".join(["ñ", "¿", "¡", "lo", "la", "el", "un", "mi", "es"]) in frase
  presenca_usa = "".join(["'","I", "the", "The", "is", "it", "you"]) in frase

  # 3. Calcular a probabilidade combinada
  if (probabilidade_verbo > 0.25 and not presenca_es and presenca_usa):
      probabilidade = 0.75  # Alta probabilidade
  elif (probabilidade_verbo > 0.15 and presenca_es):
      probabilidade = 0.5  # Média probabilidade
  else:
      probabilidade = 0.25  # Baixa probabilidade

  return probabilidade_verbo + probabilidade
```

```python
import re

# Função para normalizar texto
def normalize_text(frase):
  # Remover pontuação
  frase = re.sub(r"[^\w\s]", "", frase)
  # Converter para minúsculas
  frase = frase.lower()
  # Remover acentuações
  frase = unicodedata.normalize('NFKD', frase).encode('ascii', 'ignore').decode('utf-8')
  return frase

def word_frequency(frase):
  frequencias = {}
  for palavra in frase.split():
    if palavra in frequencias:
      frequencias[palavra] += 1
    else:
      frequencias[palavra] = 1
  return frequencias

frase = normalize_text("O rápido lobo marrom pula sobre o cachorro preguiçoso.")
frequencias = word_frequency(frase)
print(frequencias)
```

```python
import spacy

# Carregando o modelo de linguagem
nlp = spacy.load("pt_core_news_sm")

# Função para calcular a mediana dos N-grams
def ngram_median(sentence, n):
  # Tokenização e pré-processamento
  doc = nlp(sentence.lower())
  tokens = list(doc)

  # Geração de N-grams
  ngrams = []
  for i in range(len(tokens) - n + 1):
    ngrams.append(tuple(tokens[i:i + n]))

  # Cálculo da frequência
  ngram_counts = Counter(ngrams)

  # Ordenação
  sorted_ngrams = sorted(ngram_counts.items(), key=lambda item: item[1])
  #print(sorted_ngrams)
  # Mediana
  mid = len(sorted_ngrams) // 2
  #print(mid)
  #print(sorted_ngrams[mid][1])
  if len(sorted_ngrams) % 2 == 0:
    median = sorted_ngrams[mid][1]
  else:
    median = (sorted_ngrams[mid - 1][1] + sorted_ngrams[mid][1]) / 2

  return median

# Exemplo de uso
#sentence = "O rápido lobo marrom pula sobre o cachorro preguiçoso."
#n = 2  # N-gram de tamanho 2
#median_bigram = ngram_median(sentence, n)
#print("Mediana dos bigrams:", median_bigram)

```

```python
# a entrada é o vetor pre_padroes e a saída desse passo deverá ser "padrões"
import re
#from unicodedata import category

def tamanhoMedioFrases(texto):
  palavras = re.split("\s",texto)
  #print(palavras)
  tamanhos = [len(s) for s in palavras if len(s)>0]
  #print(tamanhos)
  soma = 0
  for t in tamanhos:
    soma=soma+t
  return soma / len(tamanhos)

def extrair_consoantes(frase):
  consoantes = re.findall(r'[¡¿ñÑ]+', frase)
  letras = ''.join(consoantes)
  #print(letras)
  return len(letras)

def conta_palavras_acento(text):
  # Palavras com acento
  accent_regex = re.compile(r"\w+[\xe0-\xff]")
  # Localiza no texto as palavras
  words = re.findall(r"\w+", text)
  count = 0
  for word in words:
    if accent_regex.match(word):
      count += 1
  return count

def mediaPalavrasAcentudas(frase):
  qtdPalavras = len(re.split("\s",frase))
  acentoPalavras = conta_palavras_acento(frase)
  valor = acentoPalavras/qtdPalavras
  #print(valor)
  return valor

# Função para normalizar texto
def normalizarFrase(frase):
  # Remover pontuação
  frase = re.sub(r"[^\w\s]", "", frase)
  # Converter para minúsculas
  frase = frase.lower()
  # Remover acentuações
  frase = unicodedata.normalize('NFKD', frase).encode('ascii', 'ignore').decode('utf-8')
  return frase

def frequenciaPalavras(frase):
  frequencias = {}
  for palavra in frase.split():
    if palavra in frequencias:
      frequencias[palavra] += 1
    else:
      frequencias[palavra] = 1
  #moda = max(frequencias.values(), key=lambda x: x)
  return frequencias


def funcao2(frase):
  return 0
#
#   crie suas funções aqui
#
#

def localizar_numero(frase):
  for caractere in frase:
      if caractere.isdigit():
        return 0
  return 1

def extraiCaracteristicas(frase):
  # frase é um vetor [ 'texto', 'lingua' ]
  texto = frase[0]
  pattern_regex = re.compile('[^\w+]', re.UNICODE)
  texto = re.sub(pattern_regex,' ',texto)
  #print(texto)
  caracteristica1=tamanhoMedioFrases(texto)
  caracteristica2=regra_1(texto)
  caracteristica3=ngram_median(texto, 3)
  # acrescente as suas funcoes no vetor padrao
  padrao = [caracteristica1, caracteristica2, caracteristica3, frase[1] ]
  return padrao


def geraPadroes(frases):
  padroes = []
  for frase in frases:
    padrao = extraiCaracteristicas(frase)
    padroes.append(padrao)
  return padroes

# converte o formato [frase classe] em
# [caracteristica_1, caracteristica_2,... caracteristica n, classe]
padroes = geraPadroes(pre_padroes)

#
# apenas para visualizacao
print(padroes)

dados = pd.DataFrame(padroes)
dados
```

<h2>Treinando o modelo com SVM</h1>
<p>Separando o conjunto de treinamento do conjunto de testes</p>

```python
from sklearn.model_selection import train_test_split
import numpy as np


#from sklearn.metrics import confusion_matrix

vet = np.array(padroes)
classes = vet[:,-1]         # classes = [p[-1] for p in padroes]
#print(classes)
padroes_sem_classe = vet[:,0:-1]
#print(padroes_sem_classe)
X_train, X_test, y_train, y_test = train_test_split(padroes_sem_classe, classes, test_size=0.25, stratify=classes)
```

Com os conjuntos separados, podemos "treinar" o modelo usando a SVM.

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

treinador = svm.SVC()  #algoritmo escolhido
modelo = treinador.fit(X_train, y_train)

#
# score com os dados de treinamento
acuracia = modelo.score(X_train, y_train)
print("Acurácia nos dados de treinamento: {:.2f}%".format(acuracia * 100))

#
# melhor avaliar com a matriz de confusão
y_pred = modelo.predict(X_train)
cm = confusion_matrix(y_train, y_pred)
print(cm)
print(classification_report(y_train, y_pred))

#
# com dados de teste que não foram usados no treinamento
print('métricas mais confiáveis')
y_pred2 = modelo.predict(X_test)
cm = confusion_matrix(y_test, y_pred2)
print(cm)
print(classification_report(y_test, y_pred2))
```
